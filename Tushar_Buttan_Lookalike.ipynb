# Feature Engineering 

```python
customer_features = df.groupby('CustomerID').agg({
    'TransactionID': 'count',       # Total transactions
    'Quantity': 'sum',              # Total quantity purchased
    'TotalValue': ['sum', 'mean'],  # Total and average transaction value
    'Category': lambda x: len(set(x)),  # Number of unique product categories
}).reset_index()
```

```python
customer_features
```

```python
customer_features.columns
```

```python
customer_features.columns = ['CustomerID', 'TotalTransactions', 'TotalQuantity', 
                             'TotalValue', 'AvgTransactionValue', 'UniqueCategories']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
scaled=sc.fit_transform(customer_features.iloc[:,1:])
```

```python
scaled
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_matrix=cosine_similarity(scaled)
```

```python
similarity_matrix
```

```python
lookalike_results = {}
for idx, customer_id in enumerate(customer_features['CustomerID']):
    # Get similarity scores for the current customer
    scores = list(enumerate(similarity_matrix[idx]))
    # Sort by score (highest to lowest) and exclude self
    top_lookalikes = sorted(scores, key=lambda x: x[1], reverse=True)[1:4]
    # Store results
    lookalike_results[customer_id] = [
        (customer_features['CustomerID'][i], score) for i, score in top_lookalikes
    ]
```

```python
lookalike_results
```

```python
lookalike_df = pd.DataFrame({
    "CustomerID": list(lookalike_results.keys())[:20],
    "Lookalikes": [lookalike_results[cust_id] for cust_id in list(lookalike_results.keys())[:20]]
})
lookalike_df.to_csv("Tushar_Lookalike.csv", index=False)
```
